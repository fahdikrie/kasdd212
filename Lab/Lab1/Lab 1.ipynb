6. Buat _DataFrame_ baru dari _DataFrame_ awal dengan __mengisi seluruh nilai null__ pada kolom _Price_ dengan nilai minimum. __Tampilkan rata-rata harga__ rumah pada _DataFrame_ baru tersebut.

```python
# Jawaban no. 6

# Menampilkan statistik DataFrame awal
print("Rata-rata awal: %f" % df_house_pl['Price'].mean())

# Mendapatkan nilai minimum pada kolom Price
min_price = df_house_pl['Price'].min()

# Menginisialisasi DataFrame baru dan mengisi nilai null dengan nilai minimum
new_df_house_pl_price = df_house_pl[['Price']].fillna(min_price)

# Menampilkan statistik DataFrame baru
print("Rata-rata baru: %f" % new_df_house_pl_price['Price'].mean())
```

Dari penemuan di atas, dapat diambil beberapa kesimpulan:
- Nilai rata-rata pada nomor 5 (setelah menghapus semua row dengan nilai `null`) sama dengan rata-rata awal. - Nilai rata-rata pada nomor 6 lebih rendah dari nomor 5. Hal ini terjadi karena pada nomor 6, harga yang bernilai `null`—yang pada nomor 5 tidak masuk perhitungan—disubstitusikan dengan nilai minimum. Tentu perlakuan ini akan berdampak pada berubahnya hasil perhitungan rata-rata.
